Since in the original histogram, a lot of pixels were under the intensity 150, I decided to brighten by multiplying those intenities by alpha, as higher intensities when multiplied would cause wraparound. The best suited alpha was 1.33, which distributed the intensity upto 200. If alpha is increased further, there is some bright spotting and the image doesnt look good.

No good alpha can be found if we want to multiply the entire image with alpha as it would cause wraparound. Hence, an intensity limit (Here 150) must be set.

# 12.Perform histogram equalization to enhance it and display the histogram. Comment on why the histogram is not flat

```python
#Histogram equalization
equ = cv2.equalizeHist(gray)
cv2_imshow(equ)
```

```python
#Plotting the equalized histogram
plt.hist(equ.ravel(),256,[0,256])
plt.xlabel('Intensity')
plt.ylabel('Pixels')
plt.title('Intensity Histogram')
plt.show()
```

The histogram is not flat as when many consecutive intensity levels dont add up to a particular pixel number, those levels are skipped and the next level is taken whose cumulative pixel number approximately matches the average. This leads to jumps and gives the histogram a jagged appearance

The image is discrete (Quantized) and so instead of infinite values between two numbers, there are finite intensity levels and finite number of pixels, like buckets. In histogram equalization we remap intensity values to approximately equalize values and not redistribute all pixel intensities to make them exactly the same. In the end, pixels originally of the same intensity will still have the same intensity.


# 13.Think of a good target histogram and do histogram specification (histogram matching)

I will use the image with gamma transform (gamma=2) for my ideal histogram, as the result looked nice.

Reference: https://medium.com/@kyawsawhtoon/a-tutorial-to-histogram-equalization-497600f270e2


```python
#original gray image
image2e=gray.copy()
hist,bins = np.histogram(image2e.flatten(),256,[0,256])
#Normalizing the histogram by dividing by no of pixels
hist=hist/499508
#Plotting the histogram and CDF
plt.plot(hist)
cdf = hist.cumsum()
plt.plot(cdf, color = 'r')
plt.xlim([0,256])
plt.title('original histogram and cdf')
plt.xlabel('intensities')
plt.ylabel('Normalized pixel numbers')
plt.legend(('histogram','cdf'), loc = 'upper left')
plt.show()
```

```python
#image2d looks nice (gamma transform with gamma=2), so i will use that histogram as ideal
#converting it to grayscale
idealgray = cv2.cvtColor(ideal, cv2.COLOR_BGR2GRAY)
hist2,bins = np.histogram(idealgray.flatten(),256,[0,256])
#Normalizing the histogram by dividing by no of pixels
hist2=hist2/499508
#Plotting the histogram and CDF
plt.plot(hist2)
cdf2 = hist2.cumsum()
plt.plot(cdf2, color = 'r')
plt.title('Ideal histogram and cdf')
plt.xlabel('intensities')
plt.ylabel('Normalized pixel numbers')
plt.xlim([0,256])
plt.legend(('histogram','cdf'), loc = 'upper left')
plt.show()
```

```python
#Rounding off the ideal cdf to nearest integers after multiplying with L-1 to get 255 discrete levels
cdfround=np.rint(cdf2*255)
plt.title('ideal quantized cdf')
plt.plot(cdfround)
```

```python
#Rounding off the original cdf to nearest integers after multiplying with L-1 to get 255 discrete levels
cdfroundori=np.rint(cdf*255)
plt.title('original quantized cdf')
plt.plot(cdfroundori)
```

```python
cdfround = cdfround.astype('int')
cdfroundori = cdfroundori.astype('int')
#Printing the original and ideal cdf numpy int arrays to check them
print(cdfround)
print(cdfroundori)
```

```python
#t1 will be our transform array which we apply after histogram equalization to get the new histogram
#It is basically the inverse of the ideal quantized CDF that we had obtained
#i.e. given the y values of the ideal cdf plot, the x values will be returned
t1=np.zeros(256)
for i in range(1,256):
  #For all values between two particular levels of the cdf, we assign the index of the cdf array 
  #for eg, for 76 to 92 elements of t1, we assign intensity value 0,which is the index of 76, for 93 to 99, we assign 1,which is the index of 93, and so on.
  #As the values correspong to the y coordinate and index to x
  for j in range(cdfround[i-1],cdfround[i]):
    t1[j]=i
#Finally give 255 value to the last element (CDF property)
t1[255]=255
print (t1)
```

```python
#Step 1: Histogram equalization by applying the original CDF.
middle=image2e.copy()
for i in range(604):
  for j in range(827):
    middle[i,j]=(cdfroundori[image2e[i,j]])
#Step 2: Applying the t1 transform on the equalized image (Inverse of ideal CDF)
final=image2e.copy()
for i in range(604):
  for j in range(827):
    final[i,j]=(t1[middle[i,j]])
#Displaying the final result and the ideal image used and showing that they look almost identical
cv2_imshow(final)
cv2_imshow(idealgray)
```

We can see that after histogram specification, our ideal image exactly matches out new image.
